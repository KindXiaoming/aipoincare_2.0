```python
import numpy as np
import matplotlib.pyplot as plt

s1 = np.load('seed_experiments/dH_explained.npy')
s2 = np.load('../quantum_2/seed_experiments/dH_explained.npy')
s3 = np.load('../quantum_3/seed_experiments/dH_explained.npy')
plt.plot(np.arange(1,6), s1, marker="o")
plt.plot(np.arange(1,6), s2, marker="o")
plt.plot(np.arange(1,6), s3, marker="o")
plt.xticks([1,2,3,4,5],fontsize=15)
plt.xlabel('index', fontsize=20)
plt.yticks(fontsize=15)
plt.ylabel('Explained Ratio', fontsize=20)

plt.yscale('log')
plt.plot([1,5],[0.01,0.01],ls="--",color="black")
plt.legend([r"$(|\psi|)$",r"$(|\psi|,|\psi_x|)$",r"$(|\psi|,|\psi_x|,|\psi_{xx}|)$"],fontsize=15)
plt.savefig('./quantum.pdf', bbox_inches="tight")
plt.savefig('./quantum.png', bbox_inches="tight")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# quantum
# 1d input
s1 = [9.96414052e-01, 3.58527472e-03, 6.64519457e-07, 8.80938085e-09, 1.43220554e-11]

# 2d input
s2 = [7.98487261e-01, 1.98155597e-01, 3.35468057e-03, 2.21994135e-06,2.41372603e-07]

# 3d input
s3 = [5.86378630e-01, 3.65796684e-01, 4.68178934e-02, 1.00482658e-03,1.96599487e-06]


L = np.arange(1,6)

plt.plot(L, s1, marker="o")
plt.plot(L, s2, marker="o")
plt.plot(L, s3, marker="o")
plt.xticks([1,2,3,4,5],fontsize=15)
plt.xlabel(r'$L$', fontsize=20)
plt.yticks(fontsize=15)
plt.ylabel('Explained Ratio', fontsize=20)

plt.yscale('log')
plt.plot([1,5],[0.01,0.01],ls="--",color="black")
plt.legend([r"$(|\psi|)$",r"$(|\psi|,|\psi_x|)$",r"$(|\psi|,|\psi_x|,|\psi_{xx}|)$"],fontsize=15)
plt.savefig('./quantum_H.pdf', bbox_inches="tight")
plt.savefig('./quantums_H.png', bbox_inches="tight")
```

```python
import numpy as np
import matplotlib.pyplot as plt

s1 = np.load('seed_experiments/dH_explained.npy')
s2 = np.load('../KdV_1d_2/seed_experiments/dH_explained.npy')
s3 = np.load('../KdV_1d_3/seed_experiments/dH_explained.npy')
plt.plot(np.arange(1,6), s1, marker="o")
plt.plot(np.arange(1,6), s2, marker="o")
plt.plot(np.arange(1,6), s3, marker="o")
plt.xticks([1,2,3,4,5],fontsize=15)
plt.xlabel('index', fontsize=20)
plt.yticks(fontsize=15)
plt.ylabel('Explained Ratio', fontsize=20)

plt.yscale('log')
plt.plot([1,5],[0.01,0.01],ls="--",color="black")
plt.legend([r"$(\phi)$",r"$(\phi,\phi_x)$",r"$(\phi,\phi_x,\phi_{xx})$"],fontsize=15)
plt.savefig('./KdV_dH.pdf', bbox_inches="tight")
plt.savefig('./KdV_dH.png', bbox_inches="tight")
```

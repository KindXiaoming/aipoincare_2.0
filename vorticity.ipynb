```python
import numpy as np
import matplotlib.pyplot as plt

num_pts = 20
num_pack = 5
num_wave = 1000

x = np.linspace(-10,10,num=2*num_pts+1)[1:-1:2][np.newaxis,:]
y = np.linspace(-10,10,num=2*num_pts+1)[1:-1:2][np.newaxis,:]

def w(x,y,sigma,mux,muy,mags):
    return mags*(1/(2*np.pi*sigma**2))*np.e**(-((x-mux)**2+(y-muy)**2)/(2*sigma**2))

def wx(x,y,sigma,mux,muy,mags):
    return -w(x,y,sigma,mux,muy,mags)*(x-mux)/sigma**2

def wy(x,y,sigma,mux,muy,mags):
    return -w(x,y,sigma,mux,muy,mags)*(y-muy)/sigma**2

def wxx(x,y,sigma,mux,muy,mags):
    return (-sigma**2+(x-mux)**2)/sigma**4*w(x,y,sigma,mux,muy,mags)

def wxy(x,y,sigma,mux,muy,mags):
    return (x-mux)*(y-muy)/sigma**4**w(x,y,sigma,mux,muy,mags)

def wyy(x,y,sigma,mux,muy,mags):
    return (-sigma**2+(y-muy)**2)/sigma**4*w(x,y,sigma,mux,muy,mags)

def wxxx(x,y,sigma,mux,muy,mags):
    return (3*(x-mux)*sigma**2-(x-mux)**3)/sigma**6*w(x,y,sigma,mux,muy,mags)

def wyyy(x,y,sigma,mux,muy,mags):
    return (3*(y-muy)*sigma**2-(y-muy)**3)/sigma**6*w(x,y,sigma,mux,muy,mags)

def wxxy(x,y,sigma,mux,muy,mags):
    return (sigma**2-(x-mux)**2)*(y-muy)/sigma**6*w(x,y,sigma,mux,muy,mags)

def wxyy(x,y,sigma,mux,muy,mags):
    return (sigma**2-(y-muy)**2)*(x-mux)/sigma**6*w(x,y,sigma,mux,muy,mags)


def ux(x,y,sigma,mux,muy,mags):
    r = np.sqrt((x-mux)**2 + (y-muy)**2)
    return -mags*(y-muy)/(2*np.pi*r)*(1-np.e**(-r**2/(2*sigma**2)))

def uxx(x,y,sigma,mux,muy,mags):
    r = np.sqrt((x-mux)**2 + (y-muy)**2)
    return -mags*(x-mux)*(y-muy)/(2*np.pi*sigma**2*r**3)*(r**2+sigma**2*(1-np.e**(r**2/(2*sigma**2))))*np.e**(-r**2/(2*sigma**2))

def uxy(x,y,sigma,mux,muy,mags):
    r = np.sqrt((x-mux)**2 + (y-muy)**2)
    return mags/(2*np.pi*sigma**2*r**3)*(-y**2*r**2+x**2*sigma**2*(1-np.e**(r**2/(2*sigma**2))))*np.e**(-r**2/(2*sigma**2))

def uy(x,y,sigma,mux,muy,mags):
    r = np.sqrt((x-mux)**2 + (y-muy)**2)
    return mags*(x-mux)/(2*np.pi*r)*(1-np.e**(-r**2/(2*sigma**2)))

def uyx(x,y,sigma,mux,muy,mags):
    r = np.sqrt((x-mux)**2 + (y-muy)**2)
    return mags/(2*np.pi*sigma**2*r**3)*(x**2*r**2+y**2*sigma**2*(-1+np.e**(r**2/(2*sigma**2))))*np.e**(-r**2/(2*sigma**2))

def uyy(x,y,sigma,mux,muy,mags):
    r = np.sqrt((x-mux)**2 + (y-muy)**2)
    return mags*(x-mux)*(y-muy)/(2*np.pi*sigma**2*r**3)*(r**2+sigma**2*(1-np.e**(r**2/(2*sigma**2))))*np.e**(-r**2/(2*sigma**2))

#### Next : construct random field

sigmas = np.array([1.5]*num_pack*num_wave)[:,np.newaxis]
#mus = np.array([0.0,0.5])
mus = (np.random.rand(num_pack*num_wave,)*6-3)[:,np.newaxis]
mags = ((np.random.rand(num_pack*num_wave,)-0.5)*40)[:,np.newaxis]

p_ = p(x,sigmas,mus,mags)
px_ = px(x,sigmas,mus,mags)
pxx_ = pxx(x,sigmas,mus,mags)
pxxx_ = pxxx(x,sigmas,mus,mags)
pxxxx_ = pxxxx(x,sigmas,mus,mags)
pxxxxx_ = pxxxx(x,sigmas,mus,mags)

p_ = np.mean(p_.reshape(num_pack, num_wave, num_pts), axis=0)
px_ = np.mean(px_.reshape(num_pack, num_wave, num_pts), axis=0)
pxx_ = np.mean(pxx_.reshape(num_pack, num_wave, num_pts), axis=0)
pxxx_ = np.mean(pxxx_.reshape(num_pack, num_wave, num_pts), axis=0)
pxxxx_ = np.mean(pxxxx_.reshape(num_pack, num_wave, num_pts), axis=0)
pxxxxx_ = np.mean(pxxxxx_.reshape(num_pack, num_wave, num_pts), axis=0)

f_ = 6*p_*px_ - pxxx_
fx_ = 6*px_*px_ + 6*p_*pxx_ - pxxxx_
fxx_ = 12*px_*pxx_ + 6*px_*pxx_ + 6*p_*pxxx_ - pxxxxx_

psi = np.concatenate([p_[:,:,np.newaxis]], axis=2)

f = 6*p_*px_ - pxxx_


#plt.plot(x,np.mean(p(x,sigmas,mus,mags), axis=0))
#plt.plot(x,np.mean(px(x,sigmas,mus,mags), axis=0))
#plt.plot(x,np.mean(pxx(x,sigmas,mus,mags), axis=0))
#plt.plot(x,np.mean(pxxx(x,sigmas,mus,mags), axis=0))
```
